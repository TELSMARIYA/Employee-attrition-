```python
df=df1.copy()
```

```python
# univariate numerical
df.hist()
plt.tight_layout()
plt.show()
```

```python
for col in df.select_dtypes(exclude='O').columns:
    print('Boxplot for the column ', col)
    sns.boxplot(data=df, x=col, color='pink')
    plt.show()
```

```python
ax=sns.countplot(df,x='quit', color='green')
for p in ax.patches:
    x=p.get_bbox().get_points()[:,0]
    y=p.get_bbox().get_points()[1,1]
    ax.annotate("{:.3g}%".format(100.*y/len(df)), (x.mean(), y), 
                 ha='center', va='bottom')
plt.show()
```

```python
# Bivariates
for col in df.select_dtypes(exclude='O').columns:
    print("Bivariate Bar Graph for Outcome vs ", col)
    df.groupby('quit')[col].mean().plot(kind='bar', color='orange')
    plt.ylabel(col)
    plt.show()
```

```python
sns.pairplot(df, hue='quit')
plt.show()
```

```python
sns.heatmap(df.corr(), annot=True, cmap='Spectral', vmax=+1, vmin=-1)
plt.show()
```

```python
df2=pd.get_dummies(df)
```

```python
X=df2.drop('quit', axis=1)
y=df2.quit
```

```python
# train test split
x_train, x_test, y_train, y_test=train_test_split(X, y, test_size=0.2, 
                                random_state=101, stratify=y)
```
